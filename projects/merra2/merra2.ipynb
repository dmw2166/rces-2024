## Access Merra2 Data with Xarray

```python
import pandas as pd
```

```python
dates = pd.date_range(start='01-01-2018', end='02-08-2019', freq='M')
```

```python
for date in dates:
    y=date.strftime('%Y')
    ym = date.strftime('%Y%m')
    url = f'https://goldsmr4.gesdisc.eosdis.nasa.gov/opendap/MERRA2_MONTHLY/M2TMNXINT.5.12.4/{y}/MERRA2_100.tavgM_2d_int_Nx.{ym}.nc4'
    print(url)
```
