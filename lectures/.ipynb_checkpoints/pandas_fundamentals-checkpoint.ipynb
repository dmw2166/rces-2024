## Pandas Series Object

why pandas? works with tabular data. heterogeneoulsy typed (strings, numbers)
anything that could be done in Excel. 
plays nicely with numpy, matplotlib, etc

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#import matplotlib_inline.backend_inline
#matplotlib_inline.backend_inline.set_matplotlib_formats('retina')
#%matplotlib inline   #probably skip this cell
```

```python
names = ['Mercury', 'Venus', 'Earth']
values = [0.3e24, 4.87e24, 5.97e24]
```

A Series represents a one-dimensional array of data. The main difference between a Series and numpy array is that a Series has an index. The index contains the labels that we use to access the data. 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html

You give it an array-like input, or a dictionary. and it creates a 1D list of data with labels. 

```python
masses = pd.Series(values, index=names)
masses
```

```python
type(masses)
```

```python
masses.shape
```

```python
masses.to_numpy()  #convert it to a numpy array
```

```python
masses.to_csv('mass.csv')
```

## Merging Data

```python
temperature = pd.Series([167, 464, 15, -65],
                        index=['Mercury', 'Venus', 'Earth', 'Mars'],
                        name='temperature')
temperature
```

```python
df.join(temperature) #by default, uses the original number of rows in df, not in temperature
```

```python
df
```

```python
df.join(temperature, how='right') #tells it to use the number of rows in temperature
```

```python
df
```

```python
new_df = df.join(temperature, how='right')
```

```python
new_df
```
