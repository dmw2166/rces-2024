## Drawing into Axes

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x)
z = np.sin(6*x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)   #if you want the figure to be a part of this object, to add more to it
```

```python
plt.plot(x, y)  #if you just want it to standalone
```

```python
fig, axes = plt.subplots(figsize=(8,4), ncols=2)
ax0, ax1 = axes
ax0.plot(x, y)
ax1.plot(x, z)
```

```python
fig, axes = plt.subplots(figsize=(8,4), ncols=2)
axes[0].plot(x, y)
axes[1].plot(x, z)
```

## Labeling Plots

```python
fig, axes = plt.subplots(figsize=(8, 4), ncols=2)
ax0, ax1 = axes
ax0.plot(x, y)
ax0.set_xlabel('x')
ax0.set_ylabel('y')
ax0.set_title('x vs. y')

ax1.plot(x, z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('x vs. z')
```

## Line Style

## Labels, Ticks and Gridlines

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A complicated math function: $f(x) = \cos(x)$')

#ax.set_xticks(np.pi * np.array([-1, 0, 1]))
ax.set_xticks(np.array([-np.pi, 0, np.pi]))
ax.set_xticklabels(['$-\pi$', '0', '$\pi$'])

ax.set_yticks([-1, 0, 1])
ax.set_yticks(np.arange(-1, 1.1, 0.2), minor=True);

ax.grid(which='minor', linestyle='--')
ax.grid(which='major', linewidth=6)

ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A math function: $f(x) = \cos(x)$')  #dollar sign gets interpeted as math

#ax.set_xticks(np.pi * np.array([-1, 0, 1]))
ax.set_xticks(np.array([-np.pi, 0, np.pi]))
ax.set_xticklabels(['$-\pi$', '0', '$\pi$'])

ax.set_yticks([-1, 0, 1])
ax.set_yticks(np.arange(-1, 1.1, 0.2), minor=True);

ax.grid(which='minor', linestyle='--')
ax.grid(which='major', linewidth=6)

ax.plot(x,y)
ax.set_xlim(-np.pi,np.pi)
ax.set_ylim(-1,1)
```

## Dual Axes

```python
t = np.arange(0, 10, 0.1)
u = np.exp(t)
v = np.sin(2 * np.pi * t)
```

```python
fig, ax = plt.subplots()
ax.plot(t, u, color='tab:red')
ax.tick_params(axis='y', labelcolor='tab:red')
ax.set_ylabel('Left axis',color='tab:red')

ax2 = ax.twinx()
ax2.plot(t, v, color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.set_ylabel('Right axis',color='tab:blue')
```

```python
labels = ['first', 'second', 'third']
values = [10, 5, 30]

fig, axes = plt.subplots(figsize=(10, 5), ncols=2)
axes[0].bar(labels, values)
axes[1].barh(labels, values)
```

```python
#grouped bar chart
labels = ['first', 'second', 'third']
values = [10, 5, 30]
values2 = [20,10,60]

fig, axes = plt.subplots(figsize=(14, 5), ncols=3)
axes[0].bar(labels, values)
axes[1].barh(labels, values)

numlabels = np.arange(len(labels))   # label locations
width = 0.2
offset = 0.1
axes[2].bar(numlabels-offset,values,width,label='values')
axes[2].bar(numlabels+offset,values2,width,label='values2')
axes[2].legend()
axes[2].set_xticks(numlabels,labels)
```

## 3D Plotting Methods

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 50)
xx, yy = np.meshgrid(x, y)
f = np.cos(xx) * np.sin(yy)
f.shape
```

```python
fig, ax = plt.subplots(figsize=(12,4), ncols=2)
ax[0].imshow(f)
ax[1].imshow(f, origin='lower', interpolation='bicubic') #place the origin in upper or lower left

#imshow is good for showing images, can also do RGB and interpolation
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)
pc0 = ax[0].pcolormesh(x, y, f)
pc1 = ax[1].pcolormesh(xx, yy, f)
fig.colorbar(pc0, ax=ax[0],shrink=0.5, label = 'shrink colorbar')
fig.colorbar(pc1, ax=ax[1],aspect=2, label='wide colorbar')
```

```python
x_sm, y_sm, f_sm = xx[:10, :10], yy[:10, :10], f[:10,:10]
fig, ax = plt.subplots(figsize=(12,5), ncols=2)

ax[0].pcolormesh(x_sm, y_sm, f_sm, edgecolors='k')

ax[1].pcolormesh(x_sm, y_sm, f_sm[:-1, :-1], vmin=np.min(f_sm),vmax = np.max(f_sm), edgecolors='k')


#difference here is if the grid (x,y) is one element bigger than the data (f), then x and y 
#represent the corners of quadrilaterals that are colored in with the values in Z. 
#if grid is same shape as data, it inteprets the x,y as center points. 
```

```python
y_distorted = y_sm*(1 + 0.1*np.cos(6*x_sm))
```

```python
plt.figure(figsize=(12,6))
plt.pcolormesh(x_sm, y_distorted, f_sm[:-1, :-1], edgecolors='w')
plt.scatter(x_sm, y_distorted, c='k')
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)

ax[0].contour(x, y, f)
ax[1].contour(xx, yy, f)
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)
c0 = ax[0].contour(xx, yy, f, [0, 0.2, 0.3,0.6, 0.9])
c1 = ax[1].contour(xx, yy, f, 20)

plt.clabel(c0, fmt='%2.1f')
plt.colorbar(c1, ax=ax[1])
```

```python
fig, ax = plt.subplots(figsize=(12, 5), ncols=2)
clevels = np.arange(-1, 1, 0.2) + 0.1

cf0 = ax[0].contourf(xx, yy, f, clevels, cmap='magma', extend='both')
cf1 = ax[1].contourf(xx, yy, f, clevels, cmap='inferno', extend='both')

fig.colorbar(cf0, ax=ax[0])
fig.colorbar(cf1, ax=ax[1])
```
