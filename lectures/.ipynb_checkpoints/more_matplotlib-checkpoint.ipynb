## Drawing into Axes

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x)
z = np.sin(6*x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
plt.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize=(8,4), ncols=2)
ax0, ax1 = axes
ax0.plot(x, y)
ax1.plot(x, z)
```

```python
fig, axes = plt.subplots(figsize=(8,4), ncols=2)
axes[0].plot(x, y)
axes[1].plot(x, z)
```

## Labeling Plots

```python
fig, axes = plt.subplots(figsize=(8, 4), ncols=2)
ax0, ax1 = axes
ax0.plot(x, y)
ax0.set_xlabel('x')
ax0.set_ylabel('y')
ax0.set_title('x vs. y')

ax1.plot(x, z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('x vs. z')
```

## Line Style

```python
fig, axes = plt.subplots(figsize=(16, 5), ncols=3)
axes[0].plot(x, y, linestyle='dashed')
axes[0].plot(x, z, linestyle='--')

axes[1].plot(x, y, linestyle='dotted')
axes[1].plot(x, z, linestyle=':')

axes[2].plot(x, y, linestyle='dashdot', linewidth=20)
axes[2].plot(x, z, linestyle='-.', linewidth=0.5)
```

## Colors

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='k')
ax.plot(x, z, color='r')
```

```python
fig, axes = plt.subplots(figsize=(16, 5), ncols=3)
axes[0].plot(x, y, color='0.8')
axes[0].plot(x, z, color='0.1')

axes[1].plot(x, y, color=(0, 0, 1))
axes[1].plot(x, z, color=(0, 0.4, 0.3))

axes[2].plot(x, y, color='khaki')
axes[2].plot(x, z, color='salmon')
```

```python
fig, ax = plt.subplots(figsize=(12,7))
for factor in np.linspace(0.2, 1, 11):
    ax.plot(x, factor*y)
```

## Markers

```python
fig, axes = plt.subplots(figsize=(12, 5), ncols=2)
axes[0].plot(x[:20], y[:20], marker='.')
axes[0].plot(x[:20], z[:20], marker='o')

axes[1].plot(x[:20], z[:20], marker='^', markersize=10, markerfacecolor='r', markeredgecolor='k')
```

## Labels, Ticks and Gridlines

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'A complicated math function: $f(x) = \cos(x)$')

#ax.set_xticks(np.pi * np.array([-1, 0, 1]))
ax.set_xticks(np.array([-np.pi, 0, np.pi]))
ax.set_xticklabels([r'$-\pi$', '0', r'$\pi$'])

ax.set_yticks([-1, 0, 1])
ax.set_yticks(np.arange(-1, 1.1, 0.2), minor=True);

ax.grid(which='minor', linestyle='--')
ax.grid(which='major', linewidth=6)
```

## Dual Axes

```python
t = np.arange(0, 10, 0.1)
u = np.exp(t)
v = np.sin(2 * np.pi * t)
```

```python
fig, ax = plt.subplots()
ax.plot(t, u, color='tab:red')
ax.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax.twinx()
ax2.plot(t, v, color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')
```

## 2D Plotting Methods

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 50)
xx, yy = np.meshgrid(x, y)
f = np.cos(xx) * np.sin(yy)
f.shape
```

```python
fig, ax = plt.subplots(figsize=(12,4), ncols=2)
ax[0].imshow(f)
ax[1].imshow(f, origin='lower', interpolation='bicubic')
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)
pc0 = ax[0].pcolormesh(x, y, f)
pc1 = ax[1].pcolormesh(xx, yy, f)
fig.colorbar(pc0, ax=ax[0])
fig.colorbar(pc1, ax=ax[1])
```

```python
x_sm, y_sm, f_sm = xx[:10, :10], yy[:10, :10], f[:10,:10]
fig, ax = plt.subplots(figsize=(12,5), ncols=2)

ax[0].pcolormesh(x_sm, y_sm, f_sm, edgecolors='k')

ax[1].pcolormesh(x_sm, y_sm, f_sm[:-1, :-1], edgecolors='k')
```

```python
y_distorted = y_sm*(1 + 0.1*np.cos(6*x_sm))
```

```python
plt.figure(figsize=(12,6))
plt.pcolormesh(x_sm, y_distorted, f_sm[:-1, :-1], edgecolors='w')
plt.scatter(x_sm, y_distorted, c='k')
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)

ax[0].contour(x, y, f)
ax[1].contour(xx, yy, f)
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)
c0 = ax[0].contour(xx, yy, f, [0, 0.2, 0.3, 0.9])
c1 = ax[1].contour(xx, yy, f, 20)

plt.clabel(c0, fmt='%2.1f')
plt.colorbar(c1, ax=ax[1])
```

```python
fig, ax = plt.subplots(figsize=(12, 5), ncols=2)
clevels = np.arange(-1, 1, 0.2) + 0.1

cf0 = ax[0].contourf(xx, yy, f, clevels, cmap='magma', extend='both')
cf1 = ax[1].contourf(xx, yy, f, clevels, cmap='inferno', extend='both')

fig.colorbar(cf0, ax=ax[0])
fig.colorbar(cf1, ax=ax[1])
```
