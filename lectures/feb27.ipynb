# Drawing into Axes

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x)
z = np.sin(6*x)
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
```

```python
plt.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,4), ncols=2)
ax0,ax1 = axes
ax0.plot(x,y)
ax1.plot(x,z)
```

```python
fig,axes = plt.subplots(figsize=(8,4), ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,z)
```

# labeling plots

```python
fig,axes = plt.subplots(figsize=(12,4), ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('(a) x vs. y')


axes[1].plot(x,z)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('(b) x vs. z')
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,x,z)
```

```python
fig,ax = plt.subplots()
ax.plot(y,x,z,x)
```

```python
fig, axes = plt.subplots(figsize=(16,5),ncols=3)
axes[0].plot(x,y,linestyle='dashed')
axes[0].plot(x,z,linestyle='dotted')

axes[1].plot(x,y,linestyle='--',label='x vs y')
axes[1].plot(x,z,linestyle=':',label = 'x vs z')
axes[1].legend(loc='upper right',fontsize=22)

axes[2].plot(x,y,linestyle='dashdot', linewidth=20)
axes[2].plot(x,z,linestyle='-.',linewidth=0.5)
```

```python
fig,ax = plt.subplots()
ax.plot(x,y, color='k')
ax.plot(x,z, color='darkkhaki',label='x v z')
ax.legend()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y, color='0.8')
ax.plot(x,z, color='0.1',label='x v z')
ax.legend()
```

# markers

```python
fig,axes = plt.subplots(figsize=(12,5), ncols=2)
axes[0].plot(x[:20],y[:20], marker='.')
axes[0].plot(x[:20],z[:20], marker='o')

axes[1].plot(x[:20],z[:20], marker='^', markersize=20, markerfacecolor='r',markeredgecolor='k')
```

```python
fig,ax = plt.subplots(figsize=(12,7))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A math function: $y = \cos(x)$')

ax.set_xticks(np.array([-np.pi, 0, np.pi]))
ax.set_xticklabels(['$-\pi$', '0', '$\pi$'])

ax.set_yticks([-1,0,1])
ax.set_yticks(np.arange(-1,1.1,0.2), minor=True)

ax.grid(which='minor', linestyle='dashed')
ax.grid(which='major',linewidth=6)

ax.plot(x,y)

ax.set_xlim(-1*np.pi, 1*np.pi)
ax.set_ylim(-1,1)
```

# dual axes

```python
t = np.arange(0,10,0.1)
u = np.exp(t)
v = np.sin(2*np.pi*t) 
```

```python
fig, ax = plt.subplots()
ax.plot(t,u,color='r')
ax.tick_params(axis='y', labelcolor='r')
ax.set_ylabel('Left axis', color='r')

ax2 = ax.twinx()
ax2.plot(t,v,color='b')
ax2.tick_params(axis='y',labelcolor='b')
ax2.set_ylabel('Right axis',color='b')
```

## 2D plots

```python
import matplotlib.colors as colors

fig, ax = plt.subplots()
splot = ax.scatter(y,z,c=x, s=(100*z**2 + 5),norm=colors.LogNorm())
fig.colorbar(splot)
```

## bar plot

```python
labels = ['first','second','third']
values = [10,5,30]

fig, axes = plt.subplots(figsize=(10,5), ncols=2)
axes[0].bar(labels,values)
axes[1].barh(labels,values)
```

```python
labels = ['first','second','third']
values = [10,5,30]
values2 = [20,10,60]

fig, axes = plt.subplots(figsize=(14,5), ncols=3)
axes[0].bar(labels,values)
axes[1].barh(labels,values)

numlabels = np.arange(len(labels))
width = 0.2
offset = 0.1
axes[2].bar(numlabels-offset, values, width, label = 'values')
axes[2].bar(numlabels+offset, values2, width, label = 'values2')
axes[2].set_xticks(numlabels, labels)
axes[2].legend()
```

## 3d plots

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 50)
xx,yy = np.meshgrid(x,y)
f = np.cos(xx) * np.sin(yy)
```

```python
fig, ax = plt.subplots(figsize=(12,4), ncols=2)
ax[0].imshow(f)
ax[1].imshow(f,origin='lower', interpolation='bicubic')
```

```python
fig,ax = plt.subplots(figsize=(12,5), ncols=2)
pc0 = ax[0].pcolormesh(x,y,f)
pc1 = ax[1].pcolormesh(xx,yy,f)
fig.colorbar(pc0, shrink=1.5, label = 'shrink colorbar')
fig.colorbar(pc1, aspect=2, label = 'wide colorbar')
```

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)

clevels = np.arange(-1,1.1,0.1) + 0.1

cf0 = ax[0].contourf(xx,yy,f,clevels, cmap='magma', extend='both') 
cf1 = ax[1].contourf(x,y,f,clevels, cmap='inferno', extend='max')

fig.colorbar(cf0)
fig.colorbar(cf1)
```
